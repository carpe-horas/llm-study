```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def positionwise_ffn(x, W1, b1, W2, b2):
    h = x @ W1 + b1
    h_relu = relu(h)
    out = h_relu @ W2 + b2
    return h, h_relu, out

# Example input (batch=1, seq_len=2, d_model=3)
x = np.array([
    [1.0, -2.0, 0.5],
    [-0.5, 1.2, 0.0]
])  # shape: (2, 3)

# Fixed weights and biases
W1 = np.array([
    [0.5, -1.0, 1.2],
    [0.3, 0.7, -0.6],
    [-0.2, 0.9, 0.4]
])
b1 = np.array([0.1, 0.0, -0.2])
W2 = np.array([
    [1.1, -0.3, 0.5],
    [0.2, 0.4, -1.0],
    [-0.7, 0.8, 0.9]
])
b2 = np.array([-0.2, 0.1, 0.3])

h, h_relu, out = positionwise_ffn(x, W1, b1, W2, b2)

fig, axes = plt.subplots(1, 4, figsize=(14, 3))

data_list = [
    (x, "Input"),
    (h, "After Linear 1"),
    (h_relu, "After ReLU"),
    (out, "After Linear 2")
]

for i, (data, title) in enumerate(data_list):
    im = axes[i].imshow(data, aspect='auto')
    axes[i].set_title(title)
    axes[i].set_xlabel("Feature")
    axes[i].set_ylabel("Token")
    plt.colorbar(im, ax=axes[i], fraction=0.046, pad=0.04)

plt.tight_layout()
plt.show()
```
